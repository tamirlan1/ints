# Manhattan Distance

```python
# location latitude
def get_lat_dist(row):
    coords_1 = (row['start_lat'], 0)
    coords_2 = (row['end_lat'], 0)

    return round(great_circle(coords_1, coords_2).m)
train_orig['lat_dist'] = train_orig.apply(get_lat_dist, axis=1)
# train_orig[:5].apply(get_lat_dist, axis=1)
```

```python
# location longitude
def get_lng_dist(row):
    coords_1 = (0, row['start_lng'])
    coords_2 = (0, row['end_lng'])

    return round(great_circle(coords_1, coords_2).m)
train_orig['lng_dist'] = train_orig.apply(get_lng_dist, axis=1)
```

```python
train_orig['manhattan_dist'] = train_orig['lat_dist'] + train_orig['lng_dist']
```

```python
del train_orig['lat_dist']
del train_orig['lng_dist']
```

```python
# Split data to train and test sets
y_all = train_orig['duration']
x_all = train_orig.drop(['row_id', 'duration'], axis=1)

x_train, x_test, y_train, y_test = train_test_split(x_all, y_all, test_size=0.1, random_state=2018)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# XGBoost

```python
xgb_model = XGBRegressor(n_estimators=100, max_depth=14, learning_rate=0.16, colsample_bytree=1, subsample=0.8, gamma=1, n_jobs=10, random_state=222)
# xgb_model = XGBRegressor(n_estimators=100, max_depth=12, learning_rate=0.14, colsample_bytree=1, subsample=0.8, gamma=1, n_jobs=10, random_state=222)
xgb_model = xgb_model.fit(x_train[:3000000], y_train[:3000000])
```

```python
x_train.columns
```

```python
feat_scores = dict(zip(x_test.columns, xgb_model.feature_importances_))
sorted_x = sorted(feat_scores.items(), key=operator.itemgetter(1), reverse=True)
sorted_x
```

```python
pred_test = xgb_model.predict(x_test)
print('Test Mean Absolute Error:', mean_absolute_error(y_test, pred_test))

# xgb_model = XGBRegressor(n_estimators=100, max_depth=12, learning_rate=0.14, colsample_bytree=1, subsample=0.8, gamma=1, n_jobs=10, random_state=222)
# Test Mean Absolute Error: 197.60747605109682
```

```python
pred_train = xgb_model.predict(x_train)
print('Train Mean Absolute Error:', mean_absolute_error(y_train, pred_train))
```

### Hyperparameter Tuning

```python
params = {  
    'n_estimators': range(80, 121, 10),
    'max_depth': range(1,2),
    'learning_rate': np.arange(0.01,0.3, 0.01),
    'colsample_bytree': np.arange(0.3, 0.8, 0.05),
    'subsample': np.arange(0.3, 0.8, 0.05),
    'gamma': range(0, 10)
#     'reg_alpha': from_zero_positive,
#     "min_child_weight": from_zero_positive,
}
iters = 1

xgbreg = XGBRegressor()
```

```python
gs = RandomizedSearchCV(xgbreg, params, n_iter=iters, n_jobs=5, scoring='neg_mean_squared_error', cv=5, random_state=223)
gs.fit(x_train, y_train)
```

```python
XGBRegressor_model = gs.best_estimator_
XGBRegressor_model
```

```python
pred_test = XGBRegressor_model.predict(x_test)
print('Test Mean Absolute Error:', mean_absolute_error(y_test, pred_test))
```

```python
pred_train = XGBRegressor_model.predict(x_train)
print('Train Mean Absolute Error:', mean_absolute_error(y_train, pred_train))
```
